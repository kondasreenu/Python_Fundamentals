```python
my_list = ['Siya', 'Tiya', 'Guru', 'Daksh', 'Riya', 'Guru'] 
i = 0
while True:
  print(my_list[i])
  if (my_list[i] == 'Guru'):
    print('Found the name Guru')
    break
    print('After break statement') #observe the output
  i += 1

print('After while-loop exit')
```

```python
for i in range(10):    
  if i == 7:
    continue  
  print("The Number is :", i)
```

**#pass statement**

Python pass statement is used as a placeholder inside loops, functions, class, if-statement that is meant to be implemented later.


Break Continue Pass.png

```python
for i in range(15):
  print(f"Num : {i}")
  if i > 3:
    pass
  else:
    print(f"{i} is lessthan 3")
  print(f"Square num = {i*i}")
```

The print statement (Num : {i}) and (Squared Num : {i*i}) will be output during each iteration. The print statement ({i} is less than 3) will only be printed as long as i <=3. Once i>3, the pass statement is executed and we just exit the if/else block.
