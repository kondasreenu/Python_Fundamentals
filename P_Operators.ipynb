**# Operators**

Operators are special symbols in Python that carry out arithmetic or logical computation. 

The value that the operator operates on is called the operand.

# Operator Types

1. Arithmetic operators
    
2. Comparison (Relational) operators

3. Logical (Boolean) operators

4. Bitwise operators

5. Assignment operators

6. Special operators

# Arithmetic Operators


Arithmetic operators are used to perform mathematical operations like addition, subtraction, multiplication etc.

Operators.jpg




    + , -, *, /, %, //, **  are arithmetic operators

Example:

```python
x, y = 10, 20

#addition
print(x + y)

#subtraction(-)

#multiplication(*)

#division(/)

#modulo division (%)

#Floor Division (//)

#Exponent (**)
```

**# Comparision Operators**

Comparison operators are used to compare values. It either returns True or False according to the condition.

>, <, ==, !=, >=, <= are comparision operators


```python
a, b = 10, 20         

print(a < b)  #check a is less than b

print(a > b) #check a is greater than b

print(a == b) #check a is equal to b

print(a != b) #check a is not equal to b (!=)

print(a>=b) #check a greater than or equal to b

print(a<=b) #check a less than or equal to b
```

# Logical Operators

Logical operators are  **and, or, not** operators.

```python
a, b = True, False

#print a and b
print(a and b)

#print a or b
print(a or b)
#print not b
print(not b)
```

# Bitwise operators

Bitwise operators act on operands as if they were string of binary digits. It operates bit by bit

Bitwise AND    &   

Bitwise OR     |

Bitwise one's Complement  ~

Bitwise Ex-OR ^

Bitwise Right Shift >>

Bitwise Left Shift << 

```python
a, b = 10, 4  #10 --> 1010  4 --> 0100

#Bitwise AND
print("Bitwise AND", a & b)

#Bitwise OR
print("Bitwise OR", a | b)

#Bitwise NOT
print("Bitwise NOT",~b)

#Bitwise XOR
print("Bitwise XOR",a ^ b)

#Bitwise rightshift
print("Bitwise Left Shift",a << 2)

#Bitwise Leftshift
print("Bitwise Right Shift",a >> 2)
```

**# Assignment operators**

Assignment operators are used in Python to assign values to variables.

a = 5 is a simple assignment operator that assigns the value 5 on the right to the variable a on the left.

    =,  +=,  -=,  *=,  /=,  %=,  //=,  **=, &=,  |=,  ^=,  >>=,  <<= are Assignment operators

```python
a = 75

a += 10         #add AND
print(a)

#subtract AND (-=)
a = 75
a -= 10 
print(a)
#Multiply AND (*=)
a = 75
a *= 10 
print(a)
#Divide AND (/=)
a = 75
a /= 10 
print("/",a)
#Modulus AND (%=)
a = 75
a %= 10 
print("%", a)
#Floor Division (//=)
a = 75
a //= 10 
print("//", a)
#Exponent AND (**=)
a = 75
a **= 2 
print("**",a)



#Special case
a = 15
b = 2
print(f"{a} // {b} = ",a//b) 

a = -15
b = 2
print(f"{a} // {b} = ",a//b) 
# -9 -8 -7.5 -7 -6 ....0 
# smallest integer nearer to -7.5 --> -8 not -7


a = 15
b = -2
print(f"{a} // {b} = ",a // b) 
# -9 -8 -7.5 -7 -6 ....0 
# smallest integer nearer to -7.5 --> -8 not -7

a = -15
b = -2
print(f"{a} // {b} = ",a // b) 
```

**# Identity Operators**

**is and is not** are the identity operators in Python. 

They are used to check if two values (or variables) are located on the same part of the memory.

```python
a = 5
b = 5
print(a is b)    #5 is object created once both a and b points to same object

#check is not
```

```python
l1 = [1, 2, 3]
l2 = [1, 2, 3]
print(l1 is l2) # observe the output False not True. Not for complex variables like lists, sets, etc.
```

```python
s1 = "Data"
s2 = "Data"
print(s1 is not s2) #"Data" is object created once both s1 and s2 points to same object

s1 = "Data"
s2 = "Science"
print(s1 is not s2)
```

**# MemberShip Operators**

**in and not in** are the membership operators in Python. 
    
They are used to test whether a value or variable is found in a sequence (string, list, tuple, set and   dictionary).

```python
lst = [1, 2, 3, 4]
print(1 in lst)       #check 1 is present in a given list or not

#check 5 is present in a given list
print(5 in lst)
```

```python
d = {1: "a", 2: "b"}
print(1 in d)

d = {1: "a", 2: "b"}
print(3 in d)
```
